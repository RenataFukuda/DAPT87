## Dropping columns
> `axis=1` or

> `columns = ['name_of_column_to_drop1', 'name_of_column_to_drop2' ,...]`

### Drop columns based on condition

```python
df_droped = df.dropna(axis=1, thresh = df.shape[0]*0.6)
```

```python
dir(df_droped['drive'])
```

```python
df_droped_string = df_droped.select_dtypes(include='object')
```

```python
df_droped_string
```

```python
df_droped_string.groupby(['drive'])
```

```python
df_droped_string.columns
```

```python
df_droped_string['drive'].value_counts()
```

```python
df_droped_string.groupby(['VClass','drive'])['drive'].count().sort_index
```

```python
pd.set_option('display.max_rows', None)
```

```python
df_droped_string.groupby(['VClass','drive'])[df_droped_string['drive'].isna()]
```

```python
mask_display = df_droped_string['drive'].isna()
df_droped_string[mask_display].groupby(['VClass']).value_counts()
```

```python
mask_display = df_droped_string['drive'].isna()
df_droped_string[~(mask_display)].groupby(['VClass']).count()
```
